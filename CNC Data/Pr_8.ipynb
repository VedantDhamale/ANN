```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def dot_product(x,w,b):
    return ( x @ w ) + b
```

```python
inp = 47
h1 = 100
h2 = 100
op = 9
bias =1
lr=0.2
epochs = 1000


w1=np.random.rand(inp,h1)
w2= np.random.rand(h1,h2)
w3 = np.random.rand(h2,op)
w1,w2,w3
```

```python
data['Machining_Process'].replace(['Prep', 'Layer 1 Up', 'Layer 1 Down', 'Repositioning',
       'Layer 2 Up', 'Layer 2 Down', 'Layer 3 Up', 'Layer 3 Down', 'End'],value=[0,1,2,3,4,5,6,7,8],inplace=True)
```

```python
x = df.iloc[:,0:46]
```

```python
y = df['Machining_Process']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
y_train = np.array(y_train)
x_train = np.array(x_train)
```

```python
for epoch in range(epochs):
    h1= sigmoid(dot_product(x_train,w1,bias))
    h2 = sigmoid(dot_product(h1,w2,bias))
    op= sigmoid(dot_product(h2,w3,bias))
    op = np.array(op)
    #print(op.shape,y_train.shape)
    train= np.reshape(y_train,newshape=(120,1))
    error = train - op
    dop = error * op * (1 - op)
    dh2 = dop @ w3.T * h2 * (1- h2)
    dh1= dh2 @ w2.T * h1 * (1 - h1)

    w3 += lr * h2.T @ dop
    w2 += lr * h1.T @ dh2
    w1 += lr * x_train.T @ dh1 
```
