```python
import numpy as np
from matplotlib import pyplot as pt
```

```python
x=np.linspace(-10,10,100)
```

```python
#sigmoid
plt.title("Sigmoid Activation function")
pt.plot(x,1/(1+np.exp(-x)))
```

```python
#linear
plt.title("Linear Activation function")
pt.plot(x,x)
```

```python
#tanh
plt.title("Tanh Activation Function")
pt.plot(x,2/(1+np.exp(-2*x))-1)
```

```python
pt.plot(x,np.tanh(x))
```

```python

def BinaryFunc(x):
    y=[]
    for i in range(0,len(x)):
        if x[i]<0:
            y.append(0)
        else:
            y.append(1)
    return y

pt.title("Binary Activation Function")
pt.plot(x,BinaryFunc(x))
```

```python
#ReLU

def ReLUFunction(x):
    y=[]
    for i in x:
        if i<0:
            y.append(0)
        else:
            y.append(i)
    return y

pt.title("ReLU Activation Function")
pt.plot(x,ReLUFunction(x))
```

```python
#softmax
pt.title("SoftMax Function")
pt.plot(x,np.exp(x)/np.sum(np.exp(x)))
```

```python
#switch
pt.title("Switch Activation Function")
pt.plot(x,x/(1+np.exp(-x)))
```

```python
#softsign
pt.plot(x,x/(1+abs(x)))
pt.title("Softsign Activation Function")
```

```python
#Mish
#not correct 
pt.plot(x,x*np.tanh(1+np.log(1+np.exp(x))))
pt.title("Mish Activation Function")
```

```python
#Tanh Shrink
pt.plot(x,x-np.tanh(x))
pt.title("Tanh Shrink Activation Function")
```

```python
#SoftPlus
pt.plot(x,np.log(np.exp(x)+1))
pt.title("SoftPlus Activation Function")
```
