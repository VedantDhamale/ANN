```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.linspace(-10,10,100)
```

```python
plt.plot(x,x)
plt.xlabel("X")
plt.ylabel("X")
plt.title("Linear Activation function")
```

```python
plt.title("Sigmoid Activation function")
y=1/(1+np.exp(-x))
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,y)
```

```python
def TanhFunc(x):
    return((1-np.exp(-2*x))/(1+np.exp(-2*x)))
```

```python
plt.title("Tanh Activation Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,TanhFunc(x))
```

```python
def BinaryFunc(x):
    y=[]
    for i in range(0,len(x)):
        if x[i]<0:
            y.append(0)
        else:
            y.append(1)
    return y
```

```python
plt.title("Binary Activation Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,BinaryFunc(x))
```

```python
def ReLUFunc(x):
    y=[]
    for i in x:
        if i<0:
            y.append(0)
        else:
            y.append(i)
            
    return y
```

```python
plt.title("ReLU Activation Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,ReLUFunc(x))
```
